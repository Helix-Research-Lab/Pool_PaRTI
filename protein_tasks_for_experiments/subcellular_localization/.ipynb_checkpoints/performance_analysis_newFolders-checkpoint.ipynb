```python
def create_custom_bar_plot(categories, values, y_label, title, save_file_name=None, ascending=False, y_zoom=False):
    """
    Creates a bar plot with customized y-axis label, title, and saves the plot to a file.
    The categories are ranked based on their values in either ascending or descending order.

    Parameters:
    - categories: List of categories (x-axis)
    - values: List of values associated with the categories (y-axis values)
    - y_label: The label for the y-axis
    - title: The title of the plot
    - save_file_name: File name to save the plot image
    - ascending: Boolean flag to sort values in ascending order; default is False (descending order)
    """
    
    # Check if the lengths of categories and values match
    if len(categories) != len(values):
        raise ValueError("Length of categories and values must match")

    # Combine categories and values into a list of tuples and sort them based on values
    combined = list(zip(categories, values))
    combined_sorted = sorted(combined, key=lambda x: x[1], reverse=not ascending)

    # Unzip the sorted tuples back into categories and values
    sorted_categories, sorted_values = zip(*combined_sorted)

    # Create the bar plot
    plt.figure(figsize=(10, 6))  # You can adjust the figure size as needed
    plt.bar(sorted_categories, sorted_values, color='skyblue')  # You can change the color or make it a parameter

    # Customize the plot
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(rotation=90)  # Rotate category names for better readability if needed

    if y_zoom:
        y_max = np.max(values)
        y_min = np.min(values)
        plt.ylim([y_min*0.95, y_max*1.05])
    
    # Tight layout to prevent clipping of tick-labels
    plt.tight_layout()

    if not save_file_name == None:
        # Save the plot to a file
        plt.savefig(save_file_name, dpi=300)  # You can adjust the DPI as needed

    # Optionally display the plot
    plt.show()
```

```python

def plot_sorted_data_with_iqr_and_lines(data_dict, title=None, save_file_name=None, rotation=80, y_axis="Values"):
    """
    Creates a plot with individual data points, median, IQR, and lines between data points of the same index across categories,
    with categories sorted by the median values in descending order.

    Parameters:
    - data_dict: A dictionary where keys are categories and values are arrays of data points.
    - title: Title of the plot (optional).
    - save_file_name: Name of the file to save the plot (optional).
    - rotation: Rotation angle for category labels on the x-axis.
    - y_axis: Label for the y-axis.
    """
    plt.figure(figsize=(12, 6))

    # Calculate medians and sort categories by medians in descending order
    medians = {category: np.median(values) for category, values in data_dict.items()}
    sorted_categories = sorted(medians, key=medians.get, reverse=True)

    # Loop through each sorted category to plot
    for i, category in enumerate(sorted_categories, 1):
        values = data_dict[category]
        
        # Plot individual data points
        x_values = np.full(len(values), i)  # X-values are the same for all points in the category
        plt.scatter(x_values, values, alpha=0.6, label=category)

        # Calculate and plot median and IQR as a transparent box (using a bar plot here for simplicity)
        median_value = medians[category]
        q1, q3 = np.percentile(values, [25, 75])
        plt.bar(i, median_value, width=0.5, color='red', alpha=0.5, zorder=0)
        plt.bar(i, q3-q1, bottom=q1, width=0.3, color='blue', alpha=0.2, zorder=0)

    # Draw lines between points with the same index
    for index in range(len(next(iter(data_dict.values())))):  # Assuming all categories have the same number of data points
        points = [data_dict[category][index] for category in sorted_categories]
        plt.plot(range(1, len(sorted_categories) + 1), points, color='grey', alpha=0.5, linestyle='-', linewidth=1)

    # Set the x-ticks to correspond to the sorted category names
    plt.xticks(range(1, len(sorted_categories) + 1), sorted_categories, fontsize=14, rotation=rotation)
    plt.yticks(fontsize=14)
    plt.xlabel('\nCategories', fontsize=16)
    plt.ylabel(y_axis, fontsize=16)
    plt.grid(True, linestyle='--', linewidth=0.5, axis='y', alpha=0.7)
    if title:
        plt.title(title, fontsize=20)

    # Save the figure
    if save_file_name:
        plt.savefig(save_file_name, dpi=300, bbox_inches='tight')

    plt.show()
```

# MCC calculations

```python
import os
```

```python
import math

def calculate_mcc(tp, tn, fp, fn):
    """
    Calculates the Matthews Correlation Coefficient (MCC)

    Args:
        tp (int): True positives.
        tn (int): True negatives.
        fp (int): False positives.
        fn (int): False negatives.

    Returns:
        float: The MCC value.
    """

    #print(tp, tn, fp, fn)

    under_root = (tp + fp) * (tp + fn) * (tn + fp) * (tn + fn)

    # Avoid division by zero
    if under_root == 0:
        return 0

    mcc = (tp * tn - fp * fn) / math.sqrt(under_root)
    return mcc
```

```python
base_conf_dir = "./logs/confusion"
conf_files = os.listdir(base_conf_dir)
conf_files = [file for file in conf_files if ".ipynb" not in file and "old" not in file and "_16.txt" in file]
```

```python
dict_mccs = {}
for fileName in conf_files:
    #print(fileName)
    with open(f"{base_conf_dir}/{fileName}", "r") as file:
        lines = file.readlines()
    mccs = []
    TN, FP, FN, TP = -1, -1, -1, -1
    for line in lines:
        if "[[" in line:
            line = line.replace('[', '').replace(']', '').replace('\n', '')
            nums = line.strip().split(" ")
            TN = int(nums[0])
            FP = int(nums[-1])
        elif "]]" in line:
            line = line.replace('[', '').replace(']', '').replace('\n', '')
            nums = line.strip().split(" ")
            FN = int(nums[0])
            TP = int(nums[-1])
        
    
        if TN >= 0 and FP >= 0 and FN >= 0 and TP >= 0:
            mcc = calculate_mcc(tp = TP, 
                        tn = TN, 
                        fp = FP, 
                        fn = FN)
            #print(mcc)
            TN, FP, FN, TP = -1, -1, -1, -1
            mccs.append(mcc)
    #mccs
    condition = " ".join(fileName.split("_")[1:-1])
    dict_mccs[condition] = mccs
    
        
        
        
        
        
        
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_sorted_data_with_iqr_and_lines_horizontal(data_dict, title=None, save_file_name=None, rotation=0, x_axis="Categories", 
                                                   include_points=True, 
                                                   show_std_dev=False):
    """
    Creates a horizontal plot with individual data points colored by their index across categories, median, IQR,
    and lines between data points of the same index across categories,
    with categories sorted by the median values in descending order.

    Parameters:
    - data_dict: A dictionary where keys are categories and values are arrays of data points.
    - title: Title of the plot (optional).
    - save_file_name: Name of the file to save the plot (optional).
    - rotation: Rotation angle for category labels on the y-axis.
    - x_axis: Label for the x-axis.
    """
    plt.figure(figsize=(12, 16))

    # Calculate medians and sort categories by medians in descending order
    medians = {category: np.median(values) for category, values in data_dict.items()}
    means = {category: np.mean(values) for category, values in data_dict.items()}
    #sorted_categories = sorted(medians, key=medians.get, reverse=False)
    sorted_categories = sorted(means, key=means.get, reverse=False)

    # Define a color map for the indices, ensuring there are enough colors for the maximum number of data points in any category
    num_points = max(len(values) for values in data_dict.values())
    color_palette = plt.cm.viridis(np.linspace(0, 1, num_points))  # Using the viridis colormap

    # Loop through each sorted category to plot
    for i, category in enumerate(sorted_categories, 1):
        values = data_dict[category]

        # Plot individual data points horizontally with colors based on index
        for idx, value in enumerate(values):
            if include_points:
                plt.scatter(value, i, alpha=0.6, color=color_palette[idx])

        # Calculate and plot median and IQR as a transparent box (using a horizontal bar plot here for simplicity)
        median_value = medians[category]
        mean_value = means[category]
        q1, q3 = np.percentile(values, [25, 75])
        #plt.barh(i, median_value, height=0.5, color='red', alpha=0.5, zorder=0)
        plt.barh(i, mean_value, height=0.5, color='red', alpha=0.5, zorder=0)
        plt.barh(i, q3-q1, left=q1, height=0.25, color='blue', alpha=0.8, zorder=0)

        # Optionally calculate and plot standard deviation
        if show_std_dev:
            std_dev = np.std(values)
            plt.barh(i, 2*std_dev, left=mean_value-std_dev, height=0.1, color='green', alpha=0.5, zorder=0)

    # Draw lines between points with the same index in different categories
    for index in range(num_points):
        if include_points:
            points = [data_dict[category][index] if index < len(data_dict[category]) else None for category in sorted_categories]
            points = [point for point in points if point is not None]  # Remove None entries
            plt.plot(points, range(1, len(points) + 1), color=color_palette[index], alpha=0.5, linestyle='-', linewidth=1)

    # Set the y-ticks to correspond to the sorted category names
    plt.yticks(range(1, len(sorted_categories) + 1), sorted_categories, fontsize=14, rotation=rotation)
    plt.xticks(fontsize=14)
    plt.ylabel('\nCategories', fontsize=16)
    plt.xlabel(x_axis, fontsize=16)
    plt.grid(True, linestyle='--', linewidth=0.5, axis='x', alpha=0.7)
    if title:
        plt.title(title, fontsize=20)

    # Save the figure
    if save_file_name:
        plt.savefig(save_file_name, dpi=300, bbox_inches='tight')

    plt.show()
    return sorted_categories
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_horizontal_boxplot(data_dict, sort_by='median', title=None, save_file_name=None, rotation=0, x_axis="Values"):
    """
    Creates a horizontal box plot for each category with Q1, median, Q3, and mean marked,
    sorted by either the mean or median values in descending order (with the highest value at the top).

    Parameters:
    - data_dict: A dictionary where keys are categories and values are arrays of data points.
    - sort_by: Criterion for sorting categories ('median' or 'mean').
    - title: Title of the plot (optional).
    - save_file_name: Name of the file to save the plot (optional).
    - rotation: Rotation angle for category labels on the y-axis.
    - x_axis: Label for the x-axis.
    """
    plt.figure(figsize=(12, 16))

    # Calculate means and medians, then sort categories by the chosen criterion
    if sort_by == 'mean':
        values = {category: np.mean(data) for category, data in data_dict.items()}
    else:  # Default to median if anything else is specified
        values = {category: np.median(data) for category, data in data_dict.items()}
    
    sorted_categories = sorted(data_dict.keys(), key=lambda x: values[x], reverse=False)

    # Prepare data for boxplot in sorted order
    data = [data_dict[category] for category in sorted_categories]

    # Create the horizontal boxplot
    box = plt.boxplot(data, vert=False, patch_artist=True, meanline=True, showmeans=True)

    # Customizing colors of the boxplot
    for patch, median, mean in zip(box['boxes'], box['medians'], box['means']):
        patch.set_facecolor('lightblue')
        median.set_color('red')
        mean.set_color('green')
        mean.set_linestyle('--')

    # Set y-ticks and labels
    plt.yticks(range(1, len(sorted_categories) + 1), sorted_categories, fontsize=14, rotation=rotation)
    plt.xticks(fontsize=14)
    plt.ylabel("Pooling Methods", fontsize=16)
    plt.xlabel(x_axis, fontsize=16)
    plt.grid(True, linestyle='--', linewidth=0.5, axis='x', alpha=0.7)
    if title:
        plt.title(title, fontsize=20)

    # Adding a legend
    plt.plot([], [], color='red', label='Median')
    plt.plot([], [], color='green', linestyle='--', label='Mean')
    plt.legend(loc='lower right', fontsize=15)

    # Save the figure
    if save_file_name:
        plt.savefig(save_file_name, dpi=300, bbox_inches='tight')

    plt.show()
    return data

```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_horizontal_violin(data_dict, sort_by='median', title=None, save_file_name=None, rotation=0, x_axis="Values"):
    """
    Creates a horizontal violin plot for each category, sorted by either the mean or median values in descending order.
    This plot shows the distribution of the data along with density indications.

    Parameters:
    - data_dict: A dictionary where keys are categories and values are arrays of data points.
    - sort_by: Criterion for sorting categories ('median' or 'mean').
    - title: Title of the plot (optional).
    - save_file_name: Name of the file to save the plot (optional).
    - rotation: Rotation angle for category labels on the y-axis.
    - x_axis: Label for the x-axis.
    """
    plt.figure(figsize=(12, 16))

    # Calculate means and medians, then sort categories by the chosen criterion
    if sort_by == 'mean':
        values = {category: np.mean(data) for category, data in data_dict.items()}
    else:  # Default to median if anything else is specified
        values = {category: np.median(data) for category, data in data_dict.items()}
    
    sorted_categories = sorted(data_dict.keys(), key=lambda x: values[x], reverse=False)

    # Prepare data for violin plot in sorted order
    data = [data_dict[category] for category in sorted_categories]

    # Create the horizontal violin plot
    parts = plt.violinplot(data, vert=False, showmeans=True, showmedians=True, showextrema=True)

    # Customizing colors of the violin plot
    for pc in parts['bodies']:
        pc.set_facecolor('skyblue')
        pc.set_edgecolor('black')
        pc.set_alpha(0.7)
    
    # Additional customizations for mean and median lines
    parts['cmeans'].set_color('red')
    parts['cmedians'].set_color('green')

    # Set y-ticks and labels
    plt.yticks(range(1, len(sorted_categories) + 1), sorted_categories, fontsize=14, rotation=rotation)
    plt.xticks(fontsize=14)
    plt.ylabel(x_axis, fontsize=16)
    plt.xlabel('Values', fontsize=16)
    plt.grid(True, linestyle='--', linewidth=0.5, axis='x', alpha=0.7)
    if title:
        plt.title(title, fontsize=20)

    # Save the figure
    if save_file_name:
        plt.savefig(save_file_name, dpi=300, bbox_inches='tight')

    plt.show()
```

```python
plot_horizontal_violin(dict_mccs, sort_by='mean', title="Subcellular Localization", save_file_name=None, rotation=0, x_axis="MCC")
```
